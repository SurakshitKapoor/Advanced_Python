#### 3. Logical Operators:

used between the comparison operators, to merge or combine the final result of all the comparisons.

***and, or, not***

and -> if any of the comparison result is False, the final result is False 

T and F -> F;

F and T -> F;

F and F -> F;

T and T -> T

```python
# T and T
10 > 5 and 8 > 4
```

```python
# T and F
10 > 45 and 45 > 20
```

```python
-2 > -5 and 56 < -56 and -1 == -1 and 78 <= 100
```

```python
ans = 10 > 67 
ans2 = 780 > 100
ans and ans2
```

***or:*** if any of the comparison result is True,then final outcome is True.

F or T -> T;

T or F -> T;

T or T -> T;

F or F -> F

```python
# F or T -> T
10 > 90 or 45 > 5
```

```python
24 > -24 or 670 > 550 or 90 > 100
```

```python
# T or T -> T
# T and F -> F
24 > -24 or 670 > 550 and 90 > 100
```

```python
# F and F -> F
# F or T -> T
12 > 45 and 56 > 100 or 14 > 2
```

***not:*** it swaps or toggles the final result

```python
# T -> F
# F -> T
```

```python
not (10 > 6)
```

```python
not(12 > 45 and 5 > -5)
```

```python
12 > 67 or not(56 >= 56)
```

### 5. Bitwise Operators:

works on the bits -> data stored in form of bits -> 0/1 

```python
# 5 ->  0000000000000101
# 10 -> 0000000000001010

# left shift -> <<
5 << 1
```

```python
# 5 ->  0000000000000101

5 << 2
```

***>>*** : right shift

```python
# 5 ->  0000000000000101
# 2 ->  0000000000000010
5 >> 1
```

```python
# 5 ->  0000000000000101
# 1 ->  0000000000000001
5 >> 2
```

***&*** : bitwise and

```python
# 5 ->  0000000000000101
# 9 ->  0000000000001001

5 & 9
# and ,&

# 1 ->  0000000000000001
```

| -> bitwise or

```python
# 5 ->  0000000000000101
# 9 ->  0000000000001001

5 | 9 
# or  , |

# 13 -> 0000000000001101
```

***~*** -> bitwise not

```python
# 5 ->  0000000000000101
# -6 -> 1111111111111010

~5
```

***^*** -> XOR

if both bits are same -> return 0 

```python
# 5 ->  0000000000000101
# 6 ->  0000000000000110
5 ^ 6

# 3 ->  0000000000000011
```
