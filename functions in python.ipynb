```python

def calculate_sum(ls):
    sum = 0
    for i in ls:
        sum += i
        # sum = sum + i
    return sum

ans = calculate_sum([10,20,30, 40,50])
print(ans)
```

```python
# args : taking all the passed values in a tuple
def calculate_sum(x, y, *args):
    sum = x + y
    for i in list(args):
        sum += i

    return sum

# *list -> unpackaging the list
ans = calculate_sum(10, 20, *[1, 2, 3, 4, 5])
# same like: 
# ans = calculate_sum(10, 20, 1, 2, 3, 4, 5)

print(ans)
```

```python
# kwargs -> takes the assigned variables and values in the form of dictionary, where the variable becomes the key and values become sthe value
def calculate_sum(**kwargs):
    # print(kwargs)
    sum = 0
    for i in kwargs.values():
        sum += i

    return sum

ans = calculate_sum(a=10, b=20, c=30, d=40)
print(ans)
```
