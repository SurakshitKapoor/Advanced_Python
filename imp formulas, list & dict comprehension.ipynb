# Important formulas and, List Comprehensions in python

```python
ls = [7, 14, 21, 28, 35, 42, 49]
print(ls)
print(type(ls))
```

```python
# filtering -> removing even numbers

new_ls = filter(lambda x: x%2==1, ls)
new_ls
```

```python
list(new_ls)
```

```python
# sorting -> 

sorted_ls = sorted(ls, reverse=True)
sorted_ls
```

```python
# mapping -> modifying every value in a specific format

mapped_ls = map(lambda x: x+10, ls)
list(mapped_ls)
```

```python
from functools import reduce

reduced_ans = reduce(lambda acc, curr: acc+curr, ls)
```

```python
reduced_ans
```

### List comprehensions

a shorthand way to deal with list and perform operations in a single line of code

```python
new_ls = [ i*2 for i in ls]
new_ls
```

```python
new_ls = [ i*2 for i in ls if i%2==1]
new_ls
```

```python
new_ls = [ i*2 if i%2==1 else i*100 for i in ls ]
new_ls
```

```python
d = {"a":10, "b":20, "c":30, "d":40, "e":50, "f":60}
d
```

### Dictionary Comprehension

```python
new_d = { i:i*10 for i in ls }
new_d
```

```python
new_d2 = [ i  for i in d.values() if i>40 ]
new_d2
```

```python
new_d2 = { f'{i}':f"val is: {i*10}"  for i in d.values() if i>40 }
new_d2
```
