##### Tokens: 
the smallest fundamental unit in python. everything in python, is called a token.

1. ***Identifiers:*** the variables or named reference to a memory location
2. ***Literals:*** the static values used anywhere or assigned to a variable
3. ***Comments:*** the textual part in coding cell, which is ignored while execution, denote by #
4. ***Punctuations:*** Special symbols in python - :, ;, , etc.
5. ***Keywords:*** the reserved words in python, which cannot be use asa variable, eg - break, if, continue, def etc.
6. ***Operators:*** use to perform a operation between two values called operands
