# Loops in Python

Loops are used to perform repetitive tasks multiple times in a sequence. Once the loop finishes or the sequence is fully processed, the interpreter moves to the next line of code

```python
print("using the loops in python")
print()
print("for - loop")

print()

# for - loop
for i in range(0, 6):
    print(i)
```

this function runs from start till end-1, and variable i takes values from start till end-1 in a sequnece.
#### range(start, end, option:step)

```python
for i in range(1, 8+3):
    print("i: ", i)
```

```python
sum = 0
for i in range(1, 6):
    sum = sum + i
    # print(f"@ i={i}, sum is: {sum}")
    print("@ i=", i, "sum is: ", sum)
    
print("final sum is: ", sum)
```

```python
# while loop

# initialization
i = 0

# condition check
while i <= 10:
    print(i, end=" - ")

    # updating the variable
    i += 1
```

```python
# emp salary calculator
for i in range(0, 3):
    print("this is emp: ", i+1)
    salary = int(input("enter salary: "))
    # bonus = 0.1 * salary
    pct_age = int(input("enter promotion percentage: "))
    bonus = (pct_age / 100) * salary
    
    print("salary: ", salary)
    print("bonus: ", bonus)
    print("promoted_salary: ", salary + bonus)
    print()
    
```

```python
# emp salary calculator
def calc_salary(salary, pct_age):
    bonus = (pct_age / 100) * salary
    print("salary: ", salary)
    print("bonus: ", bonus)
    print("promoted_salary: ", salary + bonus)
    print()

for i in range(0, 2):
    print("this is emp: ", i+1)
    salary = int(input("enter salary: "))
    pct_age = int(input("enter promotion percentage: "))
    
    calc_salary(salary, pct_age)
    
    
```

```python
def calc(num):
    for i in range(0, num):
        print(i, end=' - ' )

    print("\texit")


print("using a loop inside a function")
calc(6)
```

```python
# conditional statements in loops
for i in range(0, 12):
    if i % 2==0:
        print(i, ": is even")
    else:
        # print(f"{i} : is odd")
        print(i, ": is odd")
```

```python
# conditional statements in loops
for i in range(0, 7):
    print(i)
    # breaking statement in loop
    if i == 4:
        break
    # print(i)
```

```python
# continue case
for i in range(0, 7):
    if i==4:
        continue
    print(i, end=' - ')
```

```python
# continue case
for i in range(0, 7):
    if i % 2 == 0:
        continue
    print(i, end=' - ')
```
