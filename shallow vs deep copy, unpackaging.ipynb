```python
# shallow vs deep copy
```

```python
# original list
ls = [12, 100, 45, 76]
ls2 = ls
ls2[2] = 450
print(ls)
print(ls2)
```

```python
ls = [12, 100, 45, 76]
ls2 = ls.copy()
ls2[2] = 450
print(ls)
print(ls2)
```

```python
# in case of nested lists, this copy method of lists would not works!
ls = [12, 100, [45, 48, 52, 56], 76]
ls2 = ls.copy()
ls2[2][2] = 450
print(ls)
print(ls2)
```

```python
solution for this is that, we have to create a deep copy, using a special method of python
```

```python
import copy
# original list
ls = [12, 100, [45, 48, 52, 56], 76]
# deeply copied list
ls2 = copy.deepcopy(ls)
ls2[2][2] = 450
print(ls)
print(ls2)
```
