# Datatypes in python

It's a way to categorize data/values to determine what type of value is stored in a variable or object.

```python
# integer type
x = 10
print(x)
print(type(x))
```

```python
# floating type
x = 12.567
print(x)
print(type(x))
```

```python
x = "hey"
print(x)
print(type(x))
```

```python
x = None
print(x)
print(type(x))
```

```python
x = 3 + 5j
print(x)
print(type(x))
```
