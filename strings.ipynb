###### string mathematical operations

1. ***concatenation*** -> adding two strings -> by '+'

in the case of two strings, the '+' behaves some differently.

'+' performs the merging or combining operation, instead of sum or add

```python
s1 = "SkillCircle"
s2 = "provides the"
s3 = "Data Science courses"

ans = s1 + s2 + s3
print(ans)
```

```python
s1 = "SkillCircle"
s2 = "provides the"
s3 = "Data Science courses"

ans = s1 + " - " + s2 + " - " + s3
print(ans)
```

```python
# eg1: printing all the words of string
s1 = "Delhi is the CApital of India.t"

# blank string
word = ""

# start iterating / traversing
for i in s1:
    # " " == " " -> tr
    if i == " ":
        print(word)
        word = ""
    else:
        # str + str -> concatenation
        # word = "" + "D" -> "D"
        # word = "D" + "e" -> "De"
        # ...
        # word = "Delhi" + " "
        word = word + i
        
```

```python
# count all the upper and lower case charcaters
s1 = "i AM from DELHI"

# counter
lowers = 0
uppers = 0
for i in s1:
    # print(i, end= ' - ')
    if i.islower()==True:
        # new_val = old_val + 1
        lowers = lowers + 1
    elif i == " ":
        continue
    else:
        uppers += 1
print("lowers chars : ", lowers)
print("upper chars : ", uppers)
```

```python
# count all the words in a sentence
s1 = "hello guys! how are you. i am fine"

# blank string
word = ""

# counter-> to count the words
freq = 0
 
# traverse over the sequence
for i in s1:
    if i == ' ':
        freq += 1
    else:
        word += i

# handling the last case
if word:
    freq += 1
    
```

```python
# count all the lowercase and uppercase characters from a string
s1 = "i AM from DELHI"

def func(s1):
    lowerCount = 0
    upperCount = 0
    word = ""
    for i in s1:
        if i == " ":
            if word.islower():
                lowerCount += 1
            else:
                upperCount += 1
            word = ""
        else:
            word = word + i
    if word.islower():
        lowerCount += 1
    else:
        upperCount += 1
    return lowerCount, upperCount

```

```python
# count the specific words from a string
s1 = "he is a boy, he got promoted, he stays in delhi."
substr = "he"

# initialize
count = 0
word = ""

# word = "h"
# word = "he"
# word = " "
# " " != " "
# word = ""
# start travresing all the chars's
for i in s1:
    if i != " ":
        word = word + i
    else:
        if word == substr:
            count += 1
        word = ""

print(count)
```
