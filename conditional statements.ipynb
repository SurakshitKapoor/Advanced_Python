whenever we want to take the decisions in python based on the conditons, we use the conditional statements

```python
# if, if's, if-else, if-elif-else
```

```python
# ex-1:
age = 12

if age >= 18:
    print("voter is eliible")
     
print("done!")
```

```python
# if-else
```

```python
age = int(input("enter age: "))
if age >= 18:
    print("the voter is eligible to vote!")
else:
    print("the voter is not eligible to cast the vote!")

print("exit")
```

```python
# using two conitions
yoe = int(input("enter years of experience: "))
salary = int( input ("enter the salary : "))

if yoe > 5 or salary >= 75000:
    print("the person has stable income an good experinece")
else:
    print("either exp. is low or salary is not good!")

print()
print("all code run successfulll!")
```

```python
# using three conditons
yoe = int(input("enter years of experience: "))
salary = int( input ("enter the salary : "))
rating = int(input ("enter employee rating: "))

if yoe > 5 or salary >= 75000 and rating >= 3:
    print("the person has stable income an good experinece")
else:
    print("either exp. is low or salary is not good!")

print()
print("all code run successfulll!")
```

```python
# if-elif-else
```

```python
marks = 35

# evaulation 
if marks >= 90:
    print("Bright student!")
elif marks >= 75:
    print("Good Student!")
elif marks >= 50:
    print("Average Student!")
elif marks > 33:
    print("Needs serious improvement!")
else:
    print("Failed!")

print('exit!')
```

```python
# simple calculator

x = int(input("enter 1st value: "))
y = int (input("enter 2nd value: "))
operator = input("enter the operator: ")

if operator == '+':
    print("Addition is: ", x+y)
elif operator == '-':
    print("subtraction is : ", x-y)
elif operator == '*':
    print("product is: ", x*y)
elif operator == '/':
    print("division is: ", x/y)
else:
    print("invalid operation!")
```
