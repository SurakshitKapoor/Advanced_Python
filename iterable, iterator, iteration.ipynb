## Iterable, Iteration, and Iterator

***iterable ->*** 

an object that can be iterated over — meaning you can traverse its elements one by one (move from one element to the next).

```python
# list, strings, tuples, sets, dictionaries are iterable objects
ls = [12, 24, 60, 56]
print(ls)
```

***iteration ->***

the process of traversing a sequence by moving from one element to the next, one at a time

```python

for i in ls:
    print(i)
```

***iterator ->*** 

The iterator is an object that keeps track of the current position while iterating over an iterable.

so,

ls → iterable

Python creates an iterator from ls internally

i → current element/value, not the iterator

for loop → iteration process

```python
# doing iteration manually
```

```python
# creating the iterator from iterable object
ls = ["apple", "banana", "cherry"]
it = iter(ls)
```

```python

it
```

`it` is the iterator;

The iterator is associated with the list, but it does not point to any element until next() is called.

```python
print(next(it))

# print("hey")

print(next(it))
print(next(it))
```

```python
print(next(it))
```


`next()` returns the current element and moves to the next;

after the last element, calling `next()` raises `StopIteration`.


When using iter() and next() manually, we control when the next element is accessed. We can execute other statements in between, so it does not behave like a loop that runs in one go.
