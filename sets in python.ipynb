# Sets in python

In Python, sets are used to store unique occurence of all the passed sequential items.

**properties:**
    
1. ***mutable:*** changes are allowed
2. ***unordered:*** not sequential
3. ***heterogeneous:*** can contains values of different types
4. ***no duplicates:*** duplicates are not allowed, contains the unique values only
5. ***no indexing:***: as sets are unordered, so cannot be accessed by indices

in python, sets are very similar to the mathematical sets. So, we can apply many mathematical sets operations here as well.

creation

```python
# empty set
s1 = set()

print(s1)
print(type(s1))
```

```python
s1 = { 12, 12, 12 }
print(s1)
print(type(s1))
```

#### Mathematical Operations in sets

1. ***Union :***

combines the sets, and takes the unique (single) occurence of all items from all the sets.

```python
s1 = {10, 20, 30, 40}
s2 = {40, 50}

union = s1.union(s2)
union
```

every item will be considered atleast once, and repetition will be denied

```python
union = s1 | s2
union
```

2. ***Intersection:***

it takes the commons elements from 2 sets.


```python
s1 = {10, 20, 30, 40}
s2 = {40, 50, 20}

intersection = s1.intersection(s2)
intersection
```

```python
intersection = s1 & s2
intersection
```

```python
s1 = {10}
s2 = {40, 50, 20}

difference = s1.difference(s2)
difference
```

```python
s1 = {10, 20, 30, 40}
s2 = {40, 50}

difference = s1.difference(s2)
difference
```

```python
ans = s1 - s2
ans
```

```python
ans = s2 - s1
ans
```

```python
s1 = {10, 20, 30, 40}
s2 = {40, 50}

symm_diff = s1.symmetric_difference(s2)
symm_diff
```
