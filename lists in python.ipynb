### Concept of shallaw copy and deep copy in lists:

*first revising the things in case of primitive type values*

```python
a = 25
b = a
print(a, id(a))
print(b, id(b))

print()

a = 100
print(a, id(a))
print(b,id(b))
```

```python
# copying a list
ls = [12, 24, 36]
ls2 = ls
print(ls, id(ls))
print(ls2, id(ls2))
```

```python
ls2[1] = True

print(ls)
print(ls2)
```

```python
# using copy method:
ls2 = ls.copy()
ls2[1] = "perfect"
print(ls, id(ls))
print(ls2, id(ls2))
```

```python
# but there is a flaw:
# when we have a nested list, then copy() method fails to deeply copy 
# the nested list elements

x = [12, 24, 36, [True, False]]
y = x.copy()

y[3][0] = "good"
print(x)
print(y)
```

```python
# soln is: use a copy library deepcopy method -> 
# which deeply copies all the values of a list, inclduing nexted list as well.

import copy 
x = [12, 24, 36, [True, False]]
y = copy.deepcopy(x)

# changes in y will not reflect in x
y[3][0] = "good"
print(x)
print(y)
```
