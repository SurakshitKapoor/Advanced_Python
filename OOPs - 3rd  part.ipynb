```python

class Student:
    school = "ABC School"   # class variable

    # constructor method: gets executed when object of this class 
    def __init__(self, name):
        self.name = name    # instance variable

    #  Instance method: works on attributes 
    def show_name(self):
        print(self.name)

    # Class method: works on class attributes
    @classmethod
    def show_school(cls):
        print(cls.school)

    #  Static method: utility functions
    @staticmethod
    def greet():
        print("Welcome Student!")

```

```python

s = Student("Amit")
```

```python
s.show_name()        # Instance method
```

```python
s2 = Student("Jai")
```

```python
Student.show_school()     # Class method
```

```python
# will change the school name for s2 student only!
s2.school = "Don Bosco Public School"
```

```python
# will show updated school name 
s2.school
```
