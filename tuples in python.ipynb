# Tuples in Python

*tuples:*

very similar to list:

1. ***heterogeneous***
2. ***ordered***
3. ***allows duplicates***
 
 but, the flaw is that the tuples are ***immutable***.
means, you cannot change the tuple by any how!


```python
# creation

# 1. a blank list
t1 = tuple()
print(t1)
print(type(t1))
```

```python
# 2. again,a blank tuple
t2 = ()
print(t2)
print(type(t2))
```

```python
t3 = (12, 56)
print(t3)
print(type(t3))
```

```python
# creating a tuple with single element
t4 = (25, )
print(t4)
print(type(t4))
```

```python
t5 = ( "Aayush" , )
t5
```

```python
# email = "raj@gmail.com"
email = input("enter employee email: ")

# defining the function
def validateEmail(email):
    valid_domains = ["hdfc.com", "icici.com", "axis.com"]
    domain = email.split('@')[-1]
    print(domain)

    if domain in valid_domains:
        print("Employee is genuine")
    else:
        print("invalid person trying to login")
```

```python
# create a new-list by opening all nested lists in a single dim. list

ls = [12, 20, True, [True, False, [100, 200], True], 30]
new_ls = []

for i in ls:
    if isinstance(i, list):
    # if type(i)==list:
        new_ls.extend(i)
    else:
        new_ls.append(i)
        
new_ls
```

```python
# remove all the vowels from list -> a, e, i, o, u
ls = ['a', 'm', 'e', 'i', 'v', 'o']

# algo:
# 1. traverse on the list
# 2. check if the current item is in vowels list or not:
#     if yes, then ignore it; (vowel case)
#     else, add it into the list (not a vowel case) 

new_ls = []
for i in ls:
    if i in ('a', 'e', 'i', 'o', 'u'):
        continue
    else:
        new_ls.append(i)

new_ls
        
```
