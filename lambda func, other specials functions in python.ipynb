#### lambda functions

special functions in python, also called anonymous functions. helps to create a single line fnction defination

used to handle simple tasks only

```python
# normalfunction declaration an call
def func(x, y):
    if x > y:
        
        return "x is greater"
    else:
        return "y is greater"

ans = func(10,20)
print(ans)
```

```python
func = lambda x, y: "x is greater" if x > y else "y is greater"
func
```

```python
func(200,30)
```

#### filter, sorted, map, reduce

```python
# filter - returns the filtered result
ls = [20, 12, 90, 67, 81, 100, 25]

def func(x):
    if x % 2 == 0:
        return x
        
# ans = filter(lambda x: x % 2 != 0 , ls)
ans = filter(func, ls)

print(ans)
print(list(ans))
```

```python
# sorts the iterable
ans = sorted(ls, reverse=True)
print(ans)
```

```python

d = [ ('a', 10), ('b', 25), ('c', 15) ]
ans = sorted(d, key=lambda x: x[1])
print(ans)
```

```python
# map -> modifies each value of iterable, and returns the updated values
ls = [20, 12, 90, 67, 81, 100, 25] 

def func(x):
    return x * 2
    
ans = map(lambda x: x * 2, ls)
# ans = map(func, ls)
print(ans)
print(tuple(ans))
```

```python
reduce -> used to aggregate the values in a single value
from functools import reduce

# reduce
ls = [20, 12, 90, 67, 81, 100, 25]

ans = reduce((lambda acc, curr: acc + curr), ls)
print(ans)
```

```python
def factorial(x):
    # base case:
    if x <= 0:
        return 1

    # tasks:
    fact = x * factorial(x-1)
    return fact

ans = factorial(5)
print(ans) 
```
